## 1.3 Comandos para crear (archivos/carpetas), mover, renombrar y eliminar

```bash
touch ejemplo.txt # crear un archivo, la extesion es opcional
ls -l
```

```bash
mkdir carpeta # crear carpeta
ls -l
```

```bash
mv ejemplo.txt carpeta # comando para mover el archivo ejemplo.txt dentro de carpeta

#para fines didacticos se realiza los siguientes comandos
#
#
echo "$PWD"
ls -l 
echo "Se movio a:"
echo "$PWD/carpeta"
ls -l carpeta

# centrarse mas en el resultado:
```

```bash
cd carpeta 
echo "antes"
ls -l

################## concentrarse en este comando:

rm ejemplo.txt # comando para eliminar archivo

####################################################

echo "despues"
ls -l
```

```bash
cd ..
echo "antes de cambiar de nombre"
ls -l
################### concentrase en el siguiente comando:

mv carpeta directory

########################################################

echo "despues de cambiar el nombre"
ls -l
```

```bash
rmdir directory # comando para borrar carpetas
ls -l
```

## 1.4 Comandos para dar permisos (lectura, escritura y ejecucion)

<p style="text-align : justify" FOND FACE="calibri">Los permisos lo podemos ver cuando listamos los archivos como se observo anteriormente:</p>

```bash
ls -ltrh
```

`d rwx rwx rwx   <--- Aqui`

se repiten 3 veces `rwx`

`1º rwx`  permisos para el creador

`2º rwx`  permisos para el grupo (varios usuarios) al que pertenece la carpeta 

`3º rwx`  permisos para los demas usuarios

¿Que significan rwx?

`r` read (Lectura) `= 4`

`w` write (Escritura) `= 2`

`x`execution (ejecución) `= 1`

```bash
cd $HOME/data
ls -lthr
```

¿Que significa si esta ausente alguno de estos caracteres?

Respuesta: 

```bash
head virus.fasta
```

```bash
chmod 444 virus.fasta # comando para cambiar permisos unicamente a lectura
ls -ltrh
```

```bash
chmod 666 sra # 6 = (4 + 2) incluye lectura y escritura pero no ejecución
ls -ltrh
```

```bash
chmod 665 sra
ls -lthr
```
