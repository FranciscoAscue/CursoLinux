### Instalacion jupyter en linux

Para instalar jupyter debemos instalar primero el kernel de `Ipython`:

```pip3 install ipykernel```

Despues podemos instalar `Jupyter`:

```pip3 install jupyter```

Despues de instalar estos programas desde la ternimal de linux, compruebe que esta operativo jupyter, puede empezar a utilizar el notebook para comprobar que el kernel esta operativo junto con los demas kernels disponbles para su computadora

Para poder usar `jupyter notebook` sola basta con escribir este comando en la terminal:

<p style="text-align : justify">Despues de ejecutarlo, se puede abrir en un navegador, solo se debe copiar la direaccion que sugiere el programa, tambien puedes escribir `localhost:8888` directamente en la direccion de un navegador. Lo importante de eso es que no necesitas internet para que se ejecute</p>
